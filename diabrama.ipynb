```python
import matplotlib.pyplot as plt
import networkx as nx

# Crear el grafo dirigido
G = nx.DiGraph()

# Definir los nodos y conexiones
nodes = {
    "Usuario": "Usuario",
    "API Gateway": "API Gateway",
    "Lambda_listar_anuncios": "Lambda (listar_anuncios)",
    "Lambda_ver_anuncio": "Lambda (ver_anuncio)",
    "Lambda_crear_anuncio": "Lambda (crear_anuncio)",
    "Lambda_listar_comentarios": "Lambda (listar_comentarios)",
    "Lambda_crear_comentario": "Lambda (crear_comentario)",
    "DynamoDB_Anuncios": "DynamoDB (Anuncios)",
    "DynamoDB_Comentarios": "DynamoDB (Comentarios)"
}

edges = [
    ("Usuario", "API Gateway"),
    ("API Gateway", "Lambda_listar_anuncios"),
    ("API Gateway", "Lambda_ver_anuncio"),
    ("API Gateway", "Lambda_crear_anuncio"),
    ("API Gateway", "Lambda_listar_comentarios"),
    ("API Gateway", "Lambda_crear_comentario"),
    ("Lambda_listar_anuncios", "DynamoDB_Anuncios"),
    ("Lambda_ver_anuncio", "DynamoDB_Anuncios"),
    ("Lambda_crear_anuncio", "DynamoDB_Anuncios"),
    ("Lambda_listar_comentarios", "DynamoDB_Comentarios"),
    ("Lambda_crear_comentario", "DynamoDB_Comentarios")
]

# Agregar nodos y bordes al grafo
G.add_nodes_from(nodes.keys())
G.add_edges_from(edges)

# Posiciones de los nodos
pos = {
    "Usuario": (0, 3),
    "API Gateway": (1, 3),
    "Lambda_listar_anuncios": (2, 4),
    "Lambda_ver_anuncio": (2, 3.5),
    "Lambda_crear_anuncio": (2, 3),
    "Lambda_listar_comentarios": (2, 2.5),
    "Lambda_crear_comentario": (2, 2),
    "DynamoDB_Anuncios": (3, 3.5),
    "DynamoDB_Comentarios": (3, 2.5)
}

# Dibujar el grafo
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", edge_color="gray", node_size=3000, font_size=10, font_weight="bold", arrowsize=15)

# Mostrar el diagrama
plt.title("Arquitectura del Proyecto: API de Anuncios en AWS", fontsize=14, fontweight='bold')
plt.show()
```
